### Combine the resulted data frames with functional groups, the presence of a functional group in a compound and the relevent lable for the compound. 
### Display the functional groups that have participate in compoounds that have been identified as TTR transcriptor activator.

```python
import pandas as pd
import numpy as np

# Upload data frames 
df1 = pd.read_csv('functional_groups_from_df1_ZEROS.csv', index_col=[0])
df2 = pd.read_csv('functional_groups_from_df2_ZEROS.csv', index_col=[0])
df3 = pd.read_csv('functional_groups_from_df3_ZEROS.csv', index_col=[0])
df4 = pd.read_csv('functional_groups_from_df4_ZEROS.csv', index_col=[0])
df5 = pd.read_csv('functional_groups_from_df5_ZEROS.csv', index_col=[0])
df6 = pd.read_csv('functional_groups_from_df6_ZEROS.csv', index_col=[0])

# Create a list with the data frames 
list_of_dataframes = [df1, df2, df3, df4, df5, df6]

# Concatenate the data frames 
combined_df = pd.concat(list_of_dataframes, ignore_index=True)

# Display the data frame
print(combined_df.shape)
combined_df.head()
```

```python
# Find similar rows in 'column' 
group_by_column = 'column'

# Define the aggregations, specifying the aggregations
summarized_values = combined_df.groupby(group_by_column).agg(
    Count_target_1=('count_target_one', 'sum'),           # Sum of target 1 for each Analyzed_Column
    Count_target_0=('count_target_zero', 'sum'),          # Sum of target 2 for each Analyzed_Column
)

# Organise a descending order 
df_sorted_desc = summarized_values.sort_values(by='Count_target_0', ascending=False)

# Display the result 
print(f"\nSummarized DataFrame by '{group_by_column}':")
print(df_sorted_desc.shape)
print(df_sorted_desc.head(30))
```
