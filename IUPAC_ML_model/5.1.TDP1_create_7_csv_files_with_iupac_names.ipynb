# Create the csv files containing CID, SMILES notations and IUPAC names

```python
import pandas as pd

#Upload data
df = pd.read_csv('TTR_full_IUPAC_data.csv')

# To avoid truncation of some columns during data frame display
pd.set_option('display.max_columns', None) 

# Display the data frame
print('Shape of the data frame: ', df.shape)
df.head()
```

```python
df.info()
```

```python
df = df[['Compound_CID', 'SMILES', 'IUPAC_Name']]
df.head()
```

```python
# Drop duplicates
df = df.drop_duplicates()
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
# Rename column 'cid'
df = df.rename(index=str, columns={'Compound_CID': 'CID',
                                   'IUPAC_Name':'UPAC'})
df.head()
```

```python
# Split data frame into 7 parts
df1 = df.iloc[:50000]
df2 = df.iloc[50001:100000]
df3 = df.iloc[100001:150000]
df4 = df.iloc[150001:200000]
df5 = df.iloc[200001:250000]
df6 = df.iloc[250001:]
```

```python
# Create csv file for each part of the data frame
df1.to_csv('df1.csv')
df2.to_csv('df2.csv')
df3.to_csv('df3.csv')
df4.to_csv('df4.csv')
df5.to_csv('df5.csv')
df6.to_csv('df6.csv')
```
